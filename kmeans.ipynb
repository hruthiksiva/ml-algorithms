```python
from sklearn.cluster import KMeans
```

```python
import numpy as np
from sklearn.datasets import make_blobs
```

```python
import matplotlib.pyplot as plt
```

```python
data, actual_label = make_blobs(n_samples=300,centers=3, cluster_std=1, random_state=42)
```

```python
data
```

```python
plt.scatter(data[:,0],data[:,1])
plt.title("K Means")
plt.legend()
plt.show()
```

```python
kmeans = KMeans(n_clusters=3,random_state=40)
kmeans.fit(data)
```

```python
centroids = kmeans.cluster_centers_
labels = kmeans.labels_
```

```python
# Visualize the clustered data
plt.scatter(data[:, 0], data[:, 1], c=labels, cmap='viridis', s=30, label='Clusters')
plt.scatter(centroids[:, 0], centroids[:, 1], s=300, c='red', marker='x', label='Centroids')
plt.title("KMeans Clustering on Generated Data")
plt.legend()
plt.show()
```
