```python
from sklearn.cluster import AgglomerativeClustering
```

```python
import numpy as np
```

```python
from sklearn.datasets import load_iris
```

```python
import matplotlib.pyplot as plt
from scipy.cluster.hierarchy import dendrogram, linkage
```

```python
df = load_iris()
data = df.data
target = df.target
```

```python
data_2d = data [:,:2]
```

```python
model = AgglomerativeClustering(n_clusters = 3, linkage='ward')
labels = model.fit_predict(data_2d)
```

```python
plt.figure(figsize=(8,5))
for i in np.unique(labels):
  plt.scatter(data_2d[i == labels,0],data_2d[i == labels,1],label=f"Cluster {labels}")

plt.title('Hierarchial Clustering')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
# plt.legend()
plt.show()
```

```python
# Step 4: Generate Dendrogram
linked = linkage(data_2d, method='ward')
plt.figure(figsize=(10, 6))
dendrogram(linked, labels=target, leaf_rotation=90, leaf_font_size=10)
plt.title("Dendrogram for Iris Dataset (2 Features)")
plt.xlabel("Sample Index")
plt.ylabel("Distance")
plt.show()
```
