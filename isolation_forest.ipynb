```python
# xAssumes anomalies are sparse in the dataset.
# Efficient for large, high-dimensional datasets.
```

```python
from sklearn.ensemble import IsolationForest
```

```python
from sklearn.datasets import make_blobs
import numpy as np
```

```python
import matplotlib.pyplot as plt
```

```python
X, y = make_blobs(n_samples=300,centers=1,cluster_std=0.6, random_state=42)
```

```python
# Add some anomalies
rng = np.random.RandomState(42)
X_outliers = rng.uniform(low=-4, high=4, size=(20, 2))
X = np.vstack([X, X_outliers])
```

```python
iso = IsolationForest(n_estimators=100,contamination=0.1, random_state=42)
```

```python
predictions = iso.fit(X)
```

```python
scores = iso.decision_function(X)
pred = iso.predict(X)
```

```python
pred
```

```python
# Step 4: Visualize results
plt.scatter(X[:, 0], X[:, 1], c=pred, cmap='coolwarm', edgecolor='k')
plt.title("Isolation Forest - Anomaly Detection")
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.show()
```
